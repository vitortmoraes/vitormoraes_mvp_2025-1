# Descrição do Problema

O conjunto de informações do dataset "IBM HR Analytics: Rotatividade e Desempenho de Funcionários" é um conjunto de dados multivariado que consiste em informações sobre os fatores que levam à rotatividade de funcionários em uma empresa, e com esses dados podemos responder a diversas perguntas relacionadas as causas e fatores importantes dessa rotatividade.

Dentre as perguntas que podemos fazer a principal seria:
Quais fatores estão mais associados à saída de funcionários da empresa?

O dataset contém dados como:
*   Idade, cargo, departamento e salário
*   Satisfação no trabalho, avaliação de desempenho, tempo de empresa
*   Informações sobre promoção, gerente e histórico profissional

A variável-alvo é "Attrition" (Rotatividade), que indica se o funcionário saiu da empresa (Yes) ou permaneceu (No).

Esse conjunto de dados é um conjunto fictício criado por cientistas de dados da IBM.

## Hipóteses do Problema

Existem muitas hipóteses e análises que podemos seguir, as principais que iremos trabalhar são as seguintes:

- Funcionários mais jovens tendem a sair com mais frequência?

- Baixa satisfação no trabalho e no ambiente pode aumentar a rotatividade?

- Funcionários com salários mais baixos ou sem promoções recentes estão mais propensos a sair?

- O tempo de casa e o tempo com o gerente atual influenciam diretamente na decisão de saída?

## Tipo de Problema

Este conjunto e informações estão relacionados a necessidade de uma **classificação supervisionada**. Devido ao conjunto de características (Satisfação com o Ambiente, Envolvimento com o Trabalho e Equilíbrio entre Vida Profissional e Pessoal), o objetivo é verificar os reais motivos de desligamento e rotatividade dos funcionários através dos dados.

O que contribui para a classificação como supervisionada:

Porque ele contém:
- Variáveis de entrada (features): como idade, cargo, salário, tempo de empresa etc.
- Uma variável-alvo (label): chamada Attrition (Rotatividade), que indica se o funcionário saiu (Yes) ou permaneceu (No) na empresa.

Isso caracteriza um problema de classificação binária supervisionada, pois:
- A saída (Attrition) já está rotulada.
- O objetivo é prever uma classe (saiu ou não saiu) com base em dados conhecidos.

# Traduzindo as colunas e outras informações

Para melhor entendimento dos atributos do dataset, iremos renomear os nomes das Colunas (Atributos) transformando as palavras que inicialmente estão inglês para o Português. E também faremos isso nas opções do Atributo "Rotatividade" e nos nomes dos Departamentos.

```python
# mudando o valor das colunas de Inglês para Português.
df.rename(columns={
    'Age': 'Idade',
    'Attrition': 'Rotatividade',
    'BusinessTravel': 'Viagem_a_Trabalho',
    'DailyRate': 'Salario_Diario',
    'Department': 'Departamento',
    'DistanceFromHome': 'Distancia_de_Casa',
    'Education': 'Nivel_Educacional',
    'EducationField': 'Area_de_Educacao',
    'EmployeeCount': 'Contagem_de_Funcionarios',
    'EmployeeNumber': 'Numero_do_Funcionario',
    'EnvironmentSatisfaction': 'Satisfacao_ambiente',
    'Gender': 'Genero',
    'HourlyRate': 'Salario_Horario',
    'JobInvolvement': 'Envolvimento_no_Trabalho',
    'JobLevel': 'Nivel_do_Cargo',
    'JobRole': 'Cargo',
    'JobSatisfaction': 'Satisfacao_no_Trabalho',
    'MaritalStatus': 'Estado_Civil',
    'MonthlyIncome': 'Salario_Mensal',
    'MonthlyRate': 'Taxa_Mensal',
    'NumCompaniesWorked': 'Num_Empresas_Anteriores',
    'Over18': 'Maior_de_18',
    'OverTime': 'Hora_Extra',
    'PercentSalaryHike': 'Aumento_Salarial_Percentual',
    'PerformanceRating': 'Avaliacao_de_Desempenho',
    'RelationshipSatisfaction': 'Satisfacao_Relacionamento',
    'StandardHours': 'Horas_Padrao',
    'StockOptionLevel': 'Nivel_Opcoes_Acoes',
    'TotalWorkingYears': 'Total_Anos_Trabalhados',
    'TrainingTimesLastYear': 'Treinamentos_Ultimo_Ano',
    'WorkLifeBalance': 'Equilibrio_Vida_Trabalho',
    'YearsAtCompany': 'Anos_na_Empresa',
    'YearsInCurrentRole': 'Anos_na_Funcao',
    'YearsSinceLastPromotion': 'Anos_desde_Promocao',
    'YearsWithCurrManager': 'Anos_com_Gerente'
}, inplace=True)

#alterando as informações do atributo "Rotatividade" que tem as informações de YES ou No para SIM e NÃO
df['Rotatividade'] = df['Rotatividade'].replace({'Yes': 'Sim', 'No': 'Não'})

# dicionário de tradução dos tipos de departamentos
traducao_departamentos = {
    'Sales': 'Vendas',
    'Research & Development': 'Pesquisa e Desenvolvimento',
    'Human Resources': 'Recursos Humanos'
}

# aplicar a tradução dos departamentos
df['Departamento'] = df['Departamento'].replace(traducao_departamentos)
```

## Histograma

A distribuição de dados descreve como os valores de uma variável se espalham, ou seja, a frequência com que diferentes valores ocorrem. Entender a distribuição é crucial na análise de dados, pois revela padrões, tendências centrais, dispersão e a presença de valores atípicos (outliers). O histograma é uma ferramenta visual fundamental para representar essa distribuição, mostrando a forma dos dados, se são simétricos ou assimétricos, unimodais ou multimodais.

**As bolinhas no canto superior esquerdo do boxplot representam outliers (valores discrepantes ou atípicos) no conjunto de dados.**

*O que são outliers?*
Outliers são valores que estão muito distantes do restante dos dados. No boxplot, eles são definidos matematicamente como:

Qualquer valor abaixo de
- Q1 - 1.5 × IQR

ou acima de
- Q3 + 1.5 × IQR

Onde:
- Q1 = Primeiro quartil (25% dos dados)
- Q3 = Terceiro quartil (75% dos dados)
- IQR = Intervalo interquartílico = Q3 − Q1

No gráfico as bolinhas aparecem no grupo "Sim" (funcionários que saíram) e indicam que alguns funcionários mais velhos saíram da empresa, mesmo sendo atípico em relação à maioria desse grupo (que é mais jovem).

**Por que isso é importante?**

Outliers podem indicar:
- Casos especiais que merecem atenção (ex: funcionários experientes saindo).
- Erros de entrada de dados (menos comum, mas possível).
- Fatores não visíveis nos dados que influenciaram essas saídas (como aposentadoria, saúde etc).

# Pré-Processamento de Dados

O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.

A primeira linha abaixo, cria a variável X, que é um conjunto de dados contendo todas as colunas do dataframe df, exceto a coluna 'Rotatividade'.

Ou seja, todas as variáveis que servirão como entrada para o modelo de machine learning. O axis=1 é utilizado para indicar que a operação de remoção deve ser feita nas colunas, e não nas linhas.

A segunda linha abaixo, cria a variável y, que contém apenas a coluna 'Rotatividade' do dataframe df. Essa coluna é a variável alvo (target), ou seja, o que você deseja prever com base nas variáveis em X.

```python
# Separar features (X) e target (y)
X = df.drop('Rotatividade', axis=1)  # Features (removendo a coluna 'Rotatividade')
y = df['Rotatividade']  # Target (coluna 'Rotatividade')
```

```python
# Identificar variáveis numéricas e categóricas
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns
```

A ideia aqui é separar os dados em duas partes: uma para treinamento e outra para validação do modelo. Para isso, você pode usar a função train_test_split, onde 30% dos dados irão para o teste e 70% para o treinamento.

```python
# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Exibir as dimensões dos conjuntos
print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")
```

A separação dos dados em treinamento e teste serve para garantir que o modelo de machine learning tenha um conjunto de dados para aprender e outro para ser validado, avaliando assim sua performance. No exemplo fornecido, a função train_test_split divide o dataset em dois conjuntos:

*Explicação do código:*

- X_train e X_test: Contêm as variáveis independentes (ou atributos) do conjunto de dados. O conjunto X_train é usado para treinar o modelo e o X_test para testar o modelo em dados nunca vistos durante o treinamento.
- y_train e y_test: Contêm a variável dependente (ou rótulo) do conjunto de dados. Similar ao X_train e X_test, y_train é usado para treinar o modelo e y_test é utilizado para validar o modelo.

*Como a separação funciona:*

- test_size=0.3: Define que 30% dos dados serão utilizados para o teste e 70% para o treinamento.
- random_state=42: Garante que a divisão dos dados seja reproduzível. Com o mesmo valor de random_state, sempre que o código for executado, a divisão dos dados será a mesma.
- stratify=y: Garante que a distribuição das classes na variável de saída y (rótulo) seja preservada entre o conjunto de treino e o de teste. Isso é importante quando se tem classes desbalanceadas, para que a divisão tenha uma distribuição proporcional das classes.

**Importância da separação:**

- Treinamento: O modelo aprende com os dados de treinamento para tentar encontrar padrões e construir uma relação entre as variáveis independentes e dependentes.
- Validação: Após o treinamento, o modelo é avaliado nos dados de teste, que não foram usados durante o processo de treinamento. Isso ajuda a avaliar a generalização do modelo, ou seja, o quanto ele será capaz de prever resultados para novos dados.

## Normalização

A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.



```python
# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()
```

O MinMaxScaler é uma técnica de normalização que ajusta os valores das variáveis para um intervalo entre 0 e 1. Isso é útil quando você está usando algoritmos que são sensíveis à escala dos dados, como regressão logística ou redes neurais.

```python
# Ajustar o scaler apenas nas variáveis numéricas
X_train_numerical = X_train[numerical_cols]
X_test_numerical = X_test[numerical_cols]

# Normalizar as variáveis numéricas
X_train_numerical_normalized = scaler_norm.fit_transform(X_train_numerical)
X_test_numerical_normalized = scaler_norm.transform(X_test_numerical)

# Ajustar o LabelEncoder em todos os dados (treino + teste) para garantir que as categorias sejam consistentes
label_encoder = LabelEncoder()

# Combinar os dados de treino e teste para aprender todas as categorias
X_combined = pd.concat([X_train[categorical_cols], X_test[categorical_cols]], axis=0)

# Ajustar o LabelEncoder em todos os dados combinados
for col in categorical_cols:
    label_encoder.fit(X_combined[col])
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])

# Combinar as variáveis numéricas e categóricas normalizadas
X_train_final = pd.concat([pd.DataFrame(X_train_numerical_normalized, columns=numerical_cols), X_train[categorical_cols].reset_index(drop=True)], axis=1)
X_test_final = pd.concat([pd.DataFrame(X_test_numerical_normalized, columns=numerical_cols), X_test[categorical_cols].reset_index(drop=True)], axis=1)

# Exibir as primeiras linhas dos dados normalizados
print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
print(X_train_final.head())
```

O método fit() calcula os valores mínimo e máximo para cada variável em X_train e os armazena no scaler_norm. Depois, com transform(), ele aplica essa normalização a X_train e X_test.

**Importante:** A normalização é feita apenas com base nos dados de treino, para evitar que informações do conjunto de teste vazem para o modelo, o que pode gerar um viés nos resultados.

```python
# Ajustar o scaler apenas nas variáveis numéricas de X_train
X_train_numerical = X_train[numerical_cols]
X_test_numerical = X_test[numerical_cols]

# Padronizar as variáveis numéricas
X_train_standardized = scaler_std.fit_transform(X_train_numerical)
X_test_standardized = scaler_std.transform(X_test_numerical)
#fit_transform é usado no conjunto de treino para aprender a média e o desvio padrão das variáveis numéricas.
#transform é usado no conjunto de teste para aplicar a mesma transformação, ou seja, usa as médias e desvios padrões calculados nos dados de treino para padronizar os dados de teste.
```

Aplicamos o StandardScaler nos dados de treinamento para calcular a média e o desvio padrão, e depois usamos essas informações para padronizar tanto os dados de treino quanto os de teste.

## 4. Identificação de Outliers

Embora a padronização e normalização reduzam a influência dos **outliers**, identificar e tratar outliers pode ainda ser importante para melhorar a qualidade dos dados. Outliers podem ser erros de entrada ou podem representar dados excepcionais que devem ser analisados separadamente.

**Métodos para Identificação de Outliers:**
- IQR (Interquartile Range): Usando o intervalo interquartil (IQR) para identificar e remover valores fora de um intervalo específico.
- Z-Score: Valores que estão a mais de 3 desvios padrão da média podem ser considerados outliers.
- Boxplots: Visualização gráfica para identificar outliers de forma rápida.

```python
# Calcular o IQR
Q1 = X_train_final['Salario_Diario'].quantile(0.25)
Q3 = X_train_final['Salario_Diario'].quantile(0.75)
IQR = Q3 - Q1

# Definir limites superior e inferior para outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remover outliers
X_train_final_cleaned  = X_train_final[(X_train_final['Salario_Diario'] >= lower_limit) & (X_train_final['Salario_Diario'] <= upper_limit)]

# Depois de remover os outliers, imprimimos o número de linhas novamente
print(f"Número de linhas após remover outliers: {X_train_final_cleaned.shape[0]}")

# Exibir algumas estatísticas das variáveis antes e depois
print("\nResumo estatístico antes e depois de remover os outliers:")
print(X_train_final[['Salario_Diario']].describe())
print(X_train_final_cleaned[['Salario_Diario']].describe())
```
